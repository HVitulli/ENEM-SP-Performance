## 1. Desempenho médio por área de conhecimento

1.  As médias das notas nas diferentes áreas do ENEM (Ciências da Natureza, Ciências Humanas, Matemática, Linguagens).

2.  Qual é a área em que os candidatos de São Paulo têm o melhor desempenho médio? E a pior?

```python
# médias
media_cn = round(df['NOTA_CN'].mean(), 2)
media_ch = round(df['NOTA_CH'].mean(), 2)
media_lc = round(df['NOTA_LC'].mean(), 2)
media_mt = round(df['NOTA_MT'].mean(), 2)

# medianas
mediana_cn = round(df['NOTA_CN'].median(), 2)
mediana_ch = round(df['NOTA_CH'].median(), 2)
mediana_lc = round(df['NOTA_LC'].median(), 2)
mediana_mt = round(df['NOTA_MT'].median(), 2)
```

```python
print('Médias das provas:\n' + 'Ciências da Natureza: ' + str(media_cn) + '\n'
      +'Ciências da Humanas: ' + str(media_ch) + '\n'
       + 'Matemática: ' + str(media_mt) + '\n'
        + 'Linguaguens: ' + str(media_lc) + '\n'
      +'\n\n'
      +'Mediana das provas:\n' + 'Ciências da Natureza: ' + str(mediana_cn) + '\n'
        +'Ciências da Humanas: ' + str(mediana_ch) + '\n'
          + 'Matemática: ' + str(mediana_mt) + '\n'
            + 'Linguaguens: ' + str(mediana_lc) + '\n'
)
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(data=df, x='NOTA_MT', kde=True)
plt.title('Distribuição de Notas - Matemática (MT)')

plt.subplot(2, 2, 2)
sns.histplot(data=df, x='NOTA_CH', kde=True)
plt.title('Distribuição de Notas - Ciências Humanas (CH)')

plt.subplot(2, 2, 3)
sns.histplot(data=df, x='NOTA_LC', kde=True)
plt.title('Distribuição de Notas - Linguagens e Códigos (LC)')

plt.subplot(2, 2, 4)
sns.histplot(data=df, x='NOTA_CN', kde=True)
plt.title('Distribuição de Notas - Ciências da Natureza (CN)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(data=df[['NOTA_MT', 'NOTA_CH', 'NOTA_LC', 'NOTA_CN']], palette=paleta_cores)
plt.title('Distribuição de Notas')
plt.xticks(ticks=[0, 1, 2, 3], labels=['MT', 'CH', 'LC', 'CN'])
plt.xlabel('Disciplinas')
plt.ylabel('Notas')

plt.show()
```

Os candidatos de São Paulo apresentaram pior desempenho em média para as disciplinas relacionadas de **Ciências da Natureza** e **Ciências Humanas**, já os melhores desempenhos em média são as provas de **Matemática** e **Linguagens e Códigos**.
