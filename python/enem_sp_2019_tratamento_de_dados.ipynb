## 2. Limpeza e organização dos dados

### 2.1 Trocando nome de colunas

### 2.2 Verificando idade dos participantes

```python
enem_df['IDADE'].describe()
```

As idades mínimas e máximas apresentadas foram 2 e 91 respectivamente, considerando isto um erro, adotarei as idades mínima e máxima 15 anos e 80 anos respectivamente.

### 2.3 Verificando valores não condizentes com as colunas das notas

```python
enem_df[['NOTA_CN','NOTA_CH','NOTA_LC','NOTA_MT','NOTA_REDACAO']].describe()
```

As notas deveriam estar em um intervalo de 0 e 1000, portanto para as notas CN, CH, LC e MT será feita uma correção divindo os mesmos por 10.

```python
enem_df[['NOTA_CN','NOTA_CH','NOTA_LC','NOTA_MT','NOTA_REDACAO']].describe()
```

### 2.4 Alterando valores dos dados com legenda para melhor legibilidade

* ESTADO_CIVIL
* COR_RACA
* NACIONALIDADE
* SITUACAO_CONCLUSAO_ENSINO
* TIPO_ESCOLA
* TREINEIRO
* TIPO_LOCALIZACAO_ESCOLA
* PRESENCA_CN
* PRESENCA_CH
* PRESENCA_LC
* PRESENCA_MT
* TIPO_LINGUA
* STATUS_REDACAO
* ACESSO_INTERNET

```python
# ESTADO_CIVIL
enem_df["ESTADO_CIVIL"] = enem_df["ESTADO_CIVIL"].replace({0:"Não Informado", 1:"Solteiro(a)", 2:"Casado(a)", 3:"Divorciado(a)", 4:"Viúvo(a)"})
```

```python
# COR_RACA
enem_df["COR_RACA"] = enem_df["COR_RACA"].replace({0:"Não Informado", 1:"Branca", 2:"Preta", 3:"Parda", 4:"Amarela", 5:"Indígena "})
```

```python
# NACIONALIDADE
enem_df["NACIONALIDADE"] = enem_df["NACIONALIDADE"].replace({0:"Não Informado", 1:"Brasileiro(a)", 2:"Brasileiro(a) Naturalizado(a)", 3:"Estrangeiro(a)",
                                                             4:"Brasileiro(a) Nato(a), nascido(a) no exterior"})
```

```python
# SITUACAO_CONCLUSAO_ENSINO
enem_df["SITUACAO_CONCLUSAO_ENSINO"] = enem_df["SITUACAO_CONCLUSAO_ENSINO"].replace({1:"Concluiu Ensino Médio", 2:"Cursando último ano do Ensino Médio",
                                                                                     3:"Cursando Ensino Médio(não último ano) ", 4:"Não concluíu e não está cursando Ensino Médio"})
```

```python
# TIPO_ESCOLA
enem_df["TIPO_ESCOLA"] = enem_df["TIPO_ESCOLA"].replace({1:"Não Informado", 2:"Pública", 3:"Privada", 4:"Exterior"})
```

```python
# TREINEIRO
enem_df["TREINEIRO"] = enem_df["TREINEIRO"].replace({0:"Não", 1:"Sim"})
```

```python
# TIPO_LOCALIZACAO_ESCOLA
enem_df["TIPO_LOCALIZACAO_ESCOLA"] = enem_df["TIPO_LOCALIZACAO_ESCOLA"].replace({1:"Urbana", 2:"Rural"})
```

```python
# PRESENCA_CN
enem_df["PRESENCA_CN"] = enem_df["PRESENCA_CN"].replace({0:"Faltou", 1:"Presente", 2:"Eliminado"})
```

```python
# PRESENCA_CH
enem_df["PRESENCA_CH"] = enem_df["PRESENCA_CH"].replace({0:"Faltou", 1:"Presente", 2:"Eliminado"})
```

```python
# PRESENCA_LC
enem_df["PRESENCA_LC"] = enem_df["PRESENCA_LC"].replace({0:"Faltou", 1:"Presente", 2:"Eliminado"})
```

```python
# PRESENCA_MT
enem_df["PRESENCA_MT"] = enem_df["PRESENCA_MT"].replace({0:"Faltou", 1:"Presente", 2:"Eliminado"})
```

```python
# TIPO_LINGUA
enem_df["TIPO_LINGUA"] = enem_df["TIPO_LINGUA"].replace({0:"Inglês", 1:"Espanhol"})
```

```python
# STATUS_REDACAO
# Possui erro de contagem dos índices do próprio banco de dados
enem_df["STATUS_REDACAO"] = enem_df["STATUS_REDACAO"].replace({1:"Sem problemas", 2:"Anulada", 3:"Cópia do texto motivador", 4:"Em branco",
                                                               6:"Fuga ao tema", 7:"Não atendimento ao tipo textual", 8:"Texto Insuficiente",
                                                               9:"Parte desconectada"})
```

```python
# ACESSO_INTERNET
enem_df["ACESSO_INTERNET"] = enem_df["ACESSO_INTERNET"].replace({"A":"Não possui", "B":"Possui"})
```

### 2.5 Verificando se escolas possuem IDs únicos

```python
enem_df['CO_ESCOLA'] = enem_df['CO_ESCOLA'].astype(str)
combinacoes_unicas = enem_df[['CO_ESCOLA', 'TIPO_ESCOLA']].drop_duplicates().sort_values('CO_ESCOLA')

print(combinacoes_unicas)
```

Existem códigos de escolas iguais com tipo_escola diferentes, como visto acima no código 35000036.0

```python
# Agrupar por 'CO_ESCOLA' e contar os valores únicos de 'TIPO_ESCOLA'
contagem_tipo_escola = enem_df.groupby('CO_ESCOLA')['TIPO_ESCOLA'].nunique()

# Encontrar o valor máximo de contagem para cada 'CO_ESCOLA'
max_contagem = contagem_tipo_escola.max()

# Encontrar a 'CO_ESCOLA' correspondente ao valor máximo de contagem
escola_max = contagem_tipo_escola.idxmax()

# Substituir o valor de 'TIPO_ESCOLA' para 'Pública' onde 'CO_ESCOLA' não corresponde ao valor máximo de contagem
enem_df.loc[enem_df['CO_ESCOLA'] != escola_max, 'TIPO_ESCOLA'] = 'Pública'
enem_df
```

```python
enem_df['CO_ESCOLA'] = enem_df['CO_ESCOLA'].astype(str)
combinacoes_unicas = enem_df[['CO_ESCOLA', 'TIPO_ESCOLA']].drop_duplicates().sort_values('CO_ESCOLA')

print(combinacoes_unicas)
```

### 3.3 Adicionando nota final

Média simples das 5 notas

```python
enem_df['MEDIA_NOTAS'] = enem_df[['NOTA_CN', 'NOTA_CH', 'NOTA_LC', 'NOTA_MT', 'NOTA_REDACAO']].mean(axis=1)
enem_df
```

```python
# Adicionando faixa de notas
faixas = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
rotulos = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800', '801-900', '901-1000']

enem_df['FAIXA_MEDIA_NOTAS'] = pd.cut(enem_df['MEDIA_NOTAS'], bins=faixas, labels=rotulos, right=False)
```
